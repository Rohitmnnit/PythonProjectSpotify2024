# Platform contributes the most to the overall popularity of the top 10 songs

```python
df['Spotify Streams']=df['Spotify Streams'].astype(str).str.replace(',','').astype(float)
df['YouTube Views']=df['YouTube Views'].astype(str).str.replace(',','').astype(float)
df['TikTok Views']=df['TikTok Views'].astype(str).str.replace(',','').astype(float)
```

```python
clean_data=df.dropna(subset=['Spotify Streams', 'YouTube Views', 'TikTok Views'])
top_10_songs=clean_data.nlargest(10,'Spotify Streams')
```

```python
platform_contributions={
    'Spotify':top_10_songs['Spotify Streams'].sum(),
    'Youtube':top_10_songs['YouTube Views'].sum(),
    'TikTok':top_10_songs['TikTok Views'].sum()
}
platform_contributions_df = pd.DataFrame(list(platform_contributions.items()), columns=['Platform', 'Total Streams/Views'])
platform_contributions_df = platform_contributions_df.sort_values(by='Total Streams/Views', ascending=False)
platform_contributions_df
```

# The top 10 songs with the highest TikTok engagement (posts, likes, views)


```python
#the top 10 songs with the highest TikTok engagement (posts, likes, views)
df['TikTok Posts'] = df['TikTok Posts'].astype(str).str.replace(',', '').astype(float)
df['TikTok Likes'] = df['TikTok Likes'].astype(str).str.replace(',', '').astype(float)
df['TikTok Views'] = df['TikTok Views'].astype(str).str.replace(',', '').astype(float)
cleaned_data = df.dropna(subset=['TikTok Posts', 'TikTok Likes', 'TikTok Views'])
cleaned_data['TikTok Engagement Score'] = (
    cleaned_data['TikTok Posts'] * 0.3 +
    cleaned_data['TikTok Likes'] * 0.3 +
    cleaned_data['TikTok Views'] * 0.4
)
top_10_tiktok_songs = cleaned_data.nlargest(10, 'TikTok Engagement Score')
top_10_tiktok_songs = top_10_tiktok_songs[['Track', 'Artist', 'TikTok Posts', 'TikTok Likes', 'TikTok Views', 'TikTok Engagement Score']]
top_10_tiktok_songs
```

# The number of Spotify playlists a song is included in affect its total streams


```python
df['Spotify Streams'] = df['Spotify Streams'].astype(str).str.replace(',', '').astype(float)
df['Spotify Playlist Count'] = df['Spotify Playlist Count'].astype(str).str.replace(',', '').astype(float)
cleaned_data = df.dropna(subset=['Spotify Streams', 'Spotify Playlist Count'])
# Scatter plot to show the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Spotify Playlist Count', y='Spotify Streams', data=cleaned_data)
plt.title('Spotify Playlist Count vs. Spotify Streams')
plt.xlabel('Spotify Playlist Count')
plt.ylabel('Spotify Streams')
plt.show()
```

# The correlation between YouTube views and YouTube likes for the top 20 songs


```python
df['YouTube Views'] = df['YouTube Views'].astype(str).str.replace(',', '').astype(float)
df['YouTube Likes'] = df['YouTube Likes'].astype(str).str.replace(',', '').astype(float)
cleaned_data = df.dropna(subset=['YouTube Views', 'YouTube Likes'])
top_20_youtube_songs = cleaned_data.nlargest(20, 'YouTube Views')
correlation_matrix = top_20_youtube_songs[['YouTube Views', 'YouTube Likes']].corr()
correlation_matrix
```

# The monthly trends in song releases among the top 100 streamed songs


```python
df['Release Date'] = pd.to_datetime(df['Release Date'])  #converting into datetime format
top_100_songs = df.nlargest(100, 'Spotify Streams')

top_100_songs['Month'] = top_100_songs['Release Date'].dt.month  #extracting month
top_100_songs['Year'] = top_100_songs['Release Date'].dt.year  #extracting year

# Plot monthly trends
monthly_trends = top_100_songs.groupby(['Year', 'Month']).size().reset_index(name='Count')
plt.figure(figsize=(15, 7))
sns.lineplot(x='Month', y='Count', hue='Year', data=monthly_trends)
plt.title('Monthly Trends in Song Releases')
plt.xlabel('Month')
plt.ylabel('Number of Releases')
plt.show()
```

# Songs released in the first half of the year compare in streams to those released in the second half


```python
top_100_songs['Half'] = top_100_songs['Month'].apply(lambda x: 'First Half' if x <= 6 else 'Second Half')
first_half_streams = top_100_songs[top_100_songs['Half'] == 'First Half']['Spotify Streams'].mean()
second_half_streams = top_100_songs[top_100_songs['Half'] == 'Second Half']['Spotify Streams'].mean()
print(f'Average streams of songs released in the first half: {first_half_streams}')
print(f'Average streams of songs released in the second half: {second_half_streams}')
```
